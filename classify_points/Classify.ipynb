```python
import pandas as pd
import numpy as np
import seaborn as sns
from keras import models
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('c:/Users/nishant/classify_points/Training Data/Logistic_X_Train.csv')
df_test = pd.read_csv('c:/Users/nishant/classify_points/Test Cases/Logistic_X_Test.csv')
```

```python
y_train = pd.read_csv('c:/Users/nishant/classify_points/Training Data/Logistic_Y_Train.csv')
```

```python
sns.scatterplot(df_train['f1'],df_train['f2'],hue=y_train['label'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train, y_train, test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = models.Sequential()
model.add(Dense(4,input_dim=2,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(X_train,y_train,epochs=150,batch_size=10,validation_data=(X_test,y_test))
```

```python
plt.plot(hist.history['loss'],color='red',label='Loss')
plt.plot(hist.history['val_loss'],color='blue',label='Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(hist.history['accuracy'],color='red',label='Accuracy')
plt.plot(hist.history['val_accuracy'],color='blue',label='Validation Accuracy')
plt.legend()
plt.show()
```

```python
y_pred = model.predict_classes(df_test)
```
